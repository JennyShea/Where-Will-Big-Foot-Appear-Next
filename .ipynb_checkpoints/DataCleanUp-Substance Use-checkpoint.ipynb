```python
import pandas as pd

#import Nation Survey on Drug Use and Health, Q1-4 2020 illicit drug use, add idu column
illicit_df = pd.read_csv('Resources/Illicit.csv')
illicit_df.head()
```

```python
illicit_df["type"]="IDU"
```

```python
illicit_df.head()
```

```python
#make new dataframe with State, 18+ Usage estimate, and type of substance
illicit_df2 = illicit_df[['State','18+ Estimate']]
illicit_df2.info()
```

```python
illicit_prepped = illicit_df2.rename(columns={"18+ Estimate":"Illicit Drug Use (18+)"})
illicit_prepped.head()
```

```python
#repeat above process with alcohol dataset
alcohol_df =pd.read_csv('Resources/Alcohol.csv')
alcohol_df.head()
```

```python
alcohol_df["type"]= "alcohol"
```

```python
alcohol_df2 = alcohol_df[['State','18+ Estimate' ]]
alcohol_df2.head()
```

```python
alcohol_prepped = alcohol_df2.rename(columns={"18+ Estimate": "Alcohol Use (18+)"})
alcohol_prepped.head()
```

```python
#repeat with substance use disorder dataset
substance_df =pd.read_csv('Resources/Alcohol.csv')
substance_df.head()
```

```python
substance_df["type"]= "sud"
```

```python
substance_df2 = substance_df[['State', '18+ Estimate']]
substance_df2.info()
```

```python
substance_prepped = substance_df2.rename(columns={"18+ Estimate": "Substance Use Disorder (18+)"})
substance_prepped.head()
```

```python
nsduh1_df = pd.merge(illicit_prepped, alcohol_prepped, on = "State", how= 'left')
```

```python
nsduh_final = pd.merge(nsduh1_df, substance_prepped, on ="State", how='left' )
```

```python
nsduh_final.to_csv('Resources/NSDUH Substance Use Data.csv')
nsduh_final.to_csv('NSDUH Substance Use Data.csv')
```
