```python
#sort sightings by year
sightings_df1 = sightings_df.sort_values(by="year")

sightings_df1.head()
```

```python
#drop the sightings from before 2001
sightings_df2 = sightings_df1[sightings_df1['year']>2000]
sightings_df2
```

```python
#create a dataframe for 2001 to 2010
sightings_00s = sightings_df2[sightings_df['year']<2011]
sightings_00s

```

```python
#get the count for 2001-2010 by state
sightings_00s_state = sightings_00s.groupby('State').count()

sightings_00s_state.head()
```

```python
bf_sitings_00s = sightings_00s_state.rename(columns= {"Unnamed": "10 Count"})
```

```python
#create dataframe for 2011-2020
sightings_10s = sightings_df2[sightings_df['year']>=2001]
sightings_10s
```

```python
#get counts of sightings for 2011-2020
sightings_10s_state = sightings_10s.groupby('State').count()
sightings_10s_state
```

```python
#Merge dataframes
aggregate_df= pd.merge(pop_change_df2, missing_df, on='State', how='left')
aggregate_df.head()
```

```python
sittings_aggregate_00_df = pd.merge(aggregate_df,sightings_00s_state, on='State', how='left')
sittings_aggregate_00_df.head()
```

```python
ag_00s_rename = sittings_aggregate_00_df.rename(columns = {"State": "State", 
                                                           "Resident Population 2020 Census": "2020 Pop",
                                                           "Percent Change 2020 Census": "2020 % Change",
                                                           "Resident Population 2010 Census": "2010 Pop",
                                                           "Percent Change 2010 Census": "2010 % Change",
                                                           "Resident Population 2000 Census": "2000 Pop",
                                                           "Total Number of Missing Persons": " # of Missing Persons",
                                                           "Missing People 100,000 Residents": "Missing Per 100k",
                                                           "Unnamed":"2001-2010 Sightings", 
                                                           "latitude":"Lat Count",
                                                           "latitude.1": "Lat1 Count",
                                                           "classification": "Class Count",
                                                           "title": "Title Count",
                                                           "Date": "Date Count",
                                                           "Year": "2001-2010 Count"
                                                            })
                                           
```

```python
agg_total=pd.merge(ag_00s_rename, sightings_10s_state, on='State', how='left' )
agg_total.head()
```

```python
agg_df2=agg_total.rename(columns = {
                                     "latitude": "20 Lat",
                                     "latitude.1": "20 Lat 1", 
                                     "classification": "20 Class", 
                                     "title": "20 Title", 
                                     "Date": "20 Date", 
                                     "Year": "20 year",
                                      })
```
